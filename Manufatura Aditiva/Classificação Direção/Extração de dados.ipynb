```python
def plano(x):
    if x == 'XY' or x == 'YX':
        return 'deitado'
    elif x == 'XZ' or x == 'YZ':
        return 'ao lado'
    elif x == 'ZX' or x == 'ZY':
        return 'em pé'
    elif x == '45':
        return x
    
def plano_id(x):
    if x == 'deitado':
        return 0
    elif x == 'ao lado':
        return 1
    elif x == 'em pé':
        return 2
    elif x == '45':
        return 3
```
