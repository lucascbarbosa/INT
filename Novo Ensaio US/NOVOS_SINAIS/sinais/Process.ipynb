```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy import signal
%matplotlib inline
```

```python
listarq = os.listdir('csv/')
sinais = []

pd.set_option('display.float_format', lambda x: '%.10f' % x)    
for arq in listarq:
    df = pd.read_csv('csv/'+arq,sep=';')
    sinais.append(df['[0]'])
sinais = pd.DataFrame(sinais)
sinais = sinais.drop(columns = [0,1,2])
```

```python
sig = sinais.iloc[0,:].apply(lambda x: '.'.join(x.split(','))).apply(lambda x: float(x))
freqs, psd = signal.welch(sig)

plt.figure(figsize=(5, 4))
plt.semilogx(freqs, psd)
plt.title('PSD: power spectral density')
plt.xlabel('Frequency')
plt.ylabel('Power')
plt.tight_layout()
```
